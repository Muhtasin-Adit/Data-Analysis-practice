```python
import pandas as pd
```

```python
df = pd.read_csv("Automobile.csv")
```

```python
df.head(10)
```

```python
df.drop_duplicates()
```

Unlike using in general, we can drop rows using subset. We can thus get the unique values.

```python
df.drop_duplicates(subset=['Company'])
```

We can use subset for multiple columns.

```python
df.drop_duplicates(subset=['Company', 'body-style'])
```
