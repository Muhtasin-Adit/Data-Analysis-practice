```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('houseprice.csv', index_col=['date'], parse_dates=True)
df.sort_index(inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(6,6), dpi=100)
plt.title('Box Plot with Seaborn')
sns.boxplot(data=df, y='price')
```

```python
#a basic example of fmri dataset of seaborn with relplot
sns.set_theme()
fmri = sns.load_dataset('fmri')

sns.relplot(data=fmri, kind='line', x='timepoint', y='signal',
 col='region', hue='event', style='event')
```

```python
df1 = pd.read_csv('ADANIPORTS.csv', parse_dates=True)
```

```python
df1.head()
```

```python
df1['H-L'] = df1['High'] - df1['Low']
```

```python
df1['100MA'] = df1['Close'].rolling(100).mean()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.scatter(df1.index, df1['H-L'], df1['100MA'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```

```python
import numpy as np 
```

```python
z1 = np.linspace(0,10,100)
x1 = np.cos(4*z1)
y1 = np.sin(4*z1)
```

```python
sns.set_style('whitegrid')
ax = plt.axes(projection='3d')
ax.plot3D(x1, y1, z1)
plt.show()
```

```python
def return_z(x,y):
    return 50-(x**2+y**2)
```

```python
sns.set_style('whitegrid')
```

```python
x1,y1 = np.linspace(-5,5,50), np.linspace(-5,5,50)
```

```python
x1, y1 = np.meshgrid(x1,y1)
z1 = return_z(x1,y1)
```

```python
plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(x1,y1,z1)
plt.show()
```
