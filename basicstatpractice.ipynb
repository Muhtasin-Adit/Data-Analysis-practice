```python
# importing the necessary libraries

import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 

%matplotlib inline
sns.set_style('darkgrid')

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('pivot.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index(['Date'], inplace = True)
```

Check for distribution and skewness using plots

```python
for col in df.columns:
    sns.distplot(df[col])
    plt.show()
```

```python
# describing the data
df.describe()
```

Inference from the describe function
1. Most number of units sold were of Drug4
2. Least number of units sold were of Drug6
3. Most sales happened with Drug4 while Drug2 contributed the least.
4. Even though Drug6 sold a max of 3 units, it generated 30,000 worth of revenue. This also means that the price of each of the drugs are not the same.

```python
import numpy as np 
```

```python
drugs = ['Drug'+ str(x) for x in range(1,9)]
for drug in drugs:
    print(drug, end=' - ')
    print(np.mean(df[drug+'_Sales']/df[drug]))
```

Inference of Drug Price
1. No price fluctuation between weeks.
2. All drugs have different price.
3. Confirms our speculation that Drug6 is costly than other drugs.
4. Drug6 is the costliest and Drug2 is the cheapest.

```python
wdf = df.resample('W').mean()
```

```python
wdf.tail()
```

```python
wdf.describe()
```

Inferences from weekly analysis
1. Drug4 tops the sales.
2. Drug6 was sold the least units.
3. Drug2 performed the worst even though 9 of its units were sold.

```python
plt.figure(figsize=(6,3), dpi=100)
plt.tick_params(axis='x', rotation=60)
plt.title('Drug1 weekly sales overview')
plt.bar(x = wdf.index, height=wdf.Drug1_Sales)
```

Inferences of Drug1 weekly sales:
1. Best week was from 29/09 to 10/5. 
2. Sales were almost same during all the weeks except the last week of september.
3. Last week had made around the double progress compared to the starting week of september.

```python
plt.figure(figsize=(6,3), dpi=100)
plt.tick_params(axis='x', rotation=60)
plt.title('Drug2 weekly sales overview')
plt.bar(x=wdf.index, height=wdf.Drug2_Sales)
```

Inferences of Drug2 weekly sales:
